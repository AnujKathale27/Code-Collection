```python
import tkinter as tk
from tkinter import messagebox
```

```python
contacts = {}
# Add a new contact

def add_contact():
    name = name_entry.get().strip()
    phone = phone_entry.get().strip()
    if not name or not phone:
        messagebox.showwarning("Input Error", "Please enter both name and phone number.")
    elif name in contacts:
        messagebox.showinfo("Info", "Contact already exists.")
    else:
        contacts[name] = phone
        messagebox.showinfo("Success", f"Contact '{name}' added.")
        clear_entries()
        display_contacts()
```

```python
# Search for a contact

def search_contact():
    name = name_entry.get().strip()
    for contact in contacts:
        if contact.lower() == name.lower():
            phone_entry.delete(0, tk.END)
            phone_entry.insert(0, contacts[contact])
            messagebox.showinfo("Found", f"Found contact: {contact} - {contacts[contact]}")
            return
    messagebox.showinfo("Not Found", "Contact not found.")
```

```python
# Update existing contact
def update_contact():
    name = name_entry.get().strip()
    if name in contacts:
        phone = phone_entry.get().strip()
        contacts[name] = phone
        messagebox.showinfo("Updated", f"Contact '{name}' updated.")
        clear_entries()
        display_contacts()
    else:
        messagebox.showwarning("Not Found", "Contact does not exist.")
```

```python
# Delete contact
def delete_contact():
    name = name_entry.get().strip()
    if name in contacts:
        del contacts[name]
        messagebox.showinfo("Deleted", f"Contact '{name}' deleted.")
        clear_entries()
        display_contacts()
    else:
        messagebox.showwarning("Not Found", "Contact does not exist.")
```

```python
# Clear the entry fields

def clear_entries():
    name_entry.delete(0, tk.END)
    phone_entry.delete(0, tk.END)
```

```python
# Display all contacts

def display_contacts():
    contact_list.delete(0, tk.END)
    for name, phone in contacts.items():
        contact_list.insert(tk.END, f"{name}: {phone}")
```

```python
# UI setup
root = tk.Tk()
root.title("Phonebook App")
root.geometry("400x400")

tk.Label(root, text="Name:").pack()
name_entry = tk.Entry(root)
name_entry.pack()

tk.Label(root, text="Phone:").pack()
phone_entry = tk.Entry(root)
phone_entry.pack()
```

```python
# Buttons
tk.Button(root, text="Add Contact", command=add_contact).pack(pady=2)
tk.Button(root, text="Search Contact", command=search_contact).pack(pady=2)
tk.Button(root, text="Update Contact", command=update_contact).pack(pady=2)
tk.Button(root, text="Delete Contact", command=delete_contact).pack(pady=2)
tk.Button(root, text="Clear Fields", command=clear_entries).pack(pady=2)
```

```python
# Display List
tk.Label(root, text="All Contacts:").pack()
contact_list = tk.Listbox(root, width=50)
contact_list.pack(pady=5)
```

```python
# Start GUI loop
root.mainloop()
```
